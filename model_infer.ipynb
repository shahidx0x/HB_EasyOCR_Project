```python
import easyocr
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
# Paths
base_image_path = '/home/shahid/Desktop/EasyOCR/trainer/all_data/bn/'
user_network = '/home/shahid/.EasyOCR/user_network'
model_storage = '/home/shahid/.EasyOCR/model'

# Initialize the EasyOCR reader
reader = easyocr.Reader(
    ['bn'],
    recog_network='bn_model',
    user_network_directory=user_network,
    model_storage_directory=model_storage,
    detector=True,
    recognizer=True,
    gpu=True
)

image_filenames = [
    'image11.jpeg',
    'image22.jpeg',
]

for image_filename in image_filenames:
    image_path = os.path.join(base_image_path, image_filename)
    image = mpimg.imread(image_path)
    result = reader.readtext(image_path, detail=0)
    plt.imshow(image)
    plt.axis('on')
    plt.show()
    
    print("Recognized Text:")
    print(result)
```
